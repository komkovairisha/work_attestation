**1.1 Скачать следующие данные: kc-house-data и laptop_price**

```python
import pandas as pd
df = pd.read_csv("kc_house_data.csv", sep=',', encoding='windows-1251')
df.head(10)
```

**1.2. Изучите стоимости недвижимости**

```python
import numpy as np
```

**Анализ матрицы корреляций**

```python
import seaborn as sns
```

```python
corr_matrix = df.corr(numeric_only=True)
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm');
```

**Выводы:**
прямая связь по параметрам: стоимость, класс дома, жилые кв.метры

**Анализ по вещественным признакам**

```python
sns.jointplot(x=df['price'], y=df['sqft_living'], kind='reg');
```

**Вывод:**
Чем больше жилая площадь дома, тем выше его стоимость

```python
sns.jointplot(x=df['lat'], y=df['long'], kind='kde');
```

**Вывод:** чем выше концентрация эллипсов, тем больше сосредоточеннсть домов

**Анализ по категориальным признакам**

```python
sns.jointplot(x=df['price'], y=df['grade'], kind='reg');
```

**Выводы:** Чем выше качество конструкции дома и дизайн, тем больше его стоимость

```python
sns.boxplot(x=df['price'], y=df['grade'].astype('str'), whis=1.5);
plt.xlabel('price')
plt.ylabel('grade')
plt.title('Distribution of price by grade');
```

**Выводы** Стоимость дома с оценкой конструкции и дизайна 13 выше, чем стоимость других домов
